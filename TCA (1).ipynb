```python
plt.figure(figsize = (3,4))
count_data = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert counts to percentages
percentage_data = count_data.div(count_data.sum(axis=1), axis=0) * 100

# Step 3: Plot the stacked bar chart
ax = percentage_data.plot(kind='bar', stacked=True, color=['skyblue', 'red'])

# Step 4: Annotate bars with percentage values
for i, row in enumerate(percentage_data.values):
    cumulative = 0
    for j, val in enumerate(row):
        plt.text(i, cumulative + val/2, f'{val:.1f}%', ha='center', va='center', fontsize=10)
        cumulative += val

# Labels and title
plt.title('Churn Percentage by Senior Citizen Status')
plt.xlabel('Senior Citizen (0 = No, 1 = Yes)')
plt.ylabel('Percentage (%)')
plt.legend(title='Churn')
plt.ylim(0, 100)
plt.xticks(ticks=[0, 1], labels=['Not Senior Citizen', 'Senior Citizen'], rotation=0)

plt.tight_layout()
plt.show()
```

```python


# Mock data reconstruction after reset
data = {
    'PhoneService': ['Yes', 'No', 'Yes', 'Yes', 'No'] * 40,
    'MultipleLines': ['No', 'Yes', 'No phone service', 'Yes', 'No'] * 40,
    'InternetService': ['DSL', 'Fiber optic', 'No', 'DSL', 'Fiber optic'] * 40,
    'OnlineSecurity': ['Yes', 'No', 'No internet service', 'Yes', 'No'] * 40,
    'OnlineBackup': ['No', 'Yes', 'No internet service', 'No', 'Yes'] * 40,
    'DeviceProtection': ['Yes', 'No', 'No internet service', 'Yes', 'No'] * 40,
    'TechSupport': ['No', 'Yes', 'No internet service', 'No', 'Yes'] * 40,
    'StreamingTV': ['Yes', 'No', 'No internet service', 'Yes', 'No'] * 40,
    'StreamingMovies': ['No', 'Yes', 'No internet service', 'No', 'Yes'] * 40,
    'Churn': ['Yes', 'No', 'No', 'Yes', 'No'] * 40
}

df = pd.DataFrame(data)

# List of columns to plot
columns_to_plot = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the matplotlib figure
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 15))
axes = axes.flatten()

# Create count plots
for i, column in enumerate(columns_to_plot):
    sns.countplot(x=column, data=df, hue='Churn', ax=axes[i])
    axes[i].set_title(f'{column} vs Churn')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
